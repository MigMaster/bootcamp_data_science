Agora, para trabalhar com mais profudidade, vamos importar o matplotlib que é o que o Pandas usa para plotar gráficos.

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.ticker as ticker
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Procedimentos Hospitalares no SUS")
plt.show()
```

```python
ax = dados.plot(x="Unidade da Federação", y="2021/Mar", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Procedimentos Hospitalares no SUS")
plt.show()
```
